<u> Problem Description: </u>

Read in 5 words from the user and store it in an appropriate data structure to facilitate printing the words in reverse order. You should take into account the following requirements:

* You can use either stack, queue, or priority queue.
* You should not use `Queue` Package.
* You should implement only two functions: `push()` and `pop()` functions.
* **Justify your selection of the data structure in a comment line**. 

<u> Sample Output: </u>

image.png

```
stack=[]

for i in range(5):
  word=input('Enter a word: ')
  stack.append(word)
  print('Word '+word+'is added')

print('Words in reverse order: ')

for i in range(5):
  print(stack.pop())
```

I chose the stack data structure because it allows me to print first words in last out (which is reverse)

<u> Problem Description: </u>

Read in 5 words from the user and store it in an appropriate data structure to facilitate printing the words according to their length. You should take into account the following requirements:

* You can use either stack, queue, or priority queue.
* You should use the `Queue` package.
* You should print the words in descending order based on length (longest to shortest).
* **Justify your selection of the data structure and solution strategy in a comment line**.


<u> Sample Output: </u>

image.png


```
import queue


q=queue.PriorityQueue()

for i in range(5):
  word=input("Enter a word: ")
  q.put((-1*len(word), word))
for i in range(5):
  print(q.get()[1])

```

I chose the priority que so that I can order the words according to their length but multiplied the length by -1 so the words are ordered descendingly (for example 'home' has the length 4, it wil be ordered following the word 'be' of length 2, because 2 is smaller than 4 but when mutliplied by -1 it is the opposite)

<h5> <u> Problem Description: </u></h5>

1. You are asked to develop a class called `taskScheduler` that includes the following: 

* Three protected attributes:
  * An appropriate data structure to store the schedueled tasks. You can use either stack, queue, or priority queue. You should use the `Queue` package.  **Justify your selection of the data structure in a comment line**.
  * A list of task type.
  * A variable to count the number of completed tasks. 
* Define a funtion called `scheduleTask()` that receives the task and its type and then add the task to the scheduel and the task type to the list of task type. Each time a new task is added, function prints an appropriate message and the number of schedueled tasks. 
* Define a function called `completeTask()` that removes the task from the schedule and its type from the task type list. The tasks should be completed in order as it scheduled.
* Define a function called `report()` that prints the number of scheduled tasks and completed tasks. 

2. Declare an object of class `taskScheduler` called `printer`.
3. Schedule the following four tasks:
    * Task: print 5 A4 pages -> Type: Printing.
    * Task: scan 8 pages -> Type: Scaning.
    * Task: print 12 A4 pages -> Type: Printing.
    * Task: print 2 A3 pages -> Type: Printing.

4. Print a report of `printer`.
5. Complete two tasks of `printer`.
6. Schedule a new task (Task: scan 4 pages -> Type: Scaning).
7. Print a report of `printer`.


<u> Sample Output: </u>

image.png


```
import queue
class taskScheduler():
  _type_list=[]
  _ourQueue=queue.Queue()
  _scheduled=0
  _complete=0
  def scheduleTask(self, task, task_type):
    self._ourQueue.put(task)
    self._type_list.append(task_type)
    self._scheduled=self._scheduled+1
    print("A new task is added to the schedule")
    print("\nA total of ", self._scheduled, "are in the queue\n-----------------------------------------------------")

  def completeTask(self):
    print(self._ourQueue.get()+' is completed \n-----------------------------------------------')
    self._complete=self._complete+1
    self._scheduled=self._scheduled-1
    
  
  def report(self):
    print('report\n\nA total of ', self._scheduled, " are in the queue.\na total of ", self._complete, ' are completed.\n')
```

```
printer= taskScheduler()
printer.scheduleTask('print 5 A4 pages', 'Printing')
printer.scheduleTask('scan 8 pages', 'Scanning')
printer.scheduleTask('print 12 A4 pages', 'Printing')
printer.scheduleTask('print 2 A3 pages', 'Printing')
printer.report()
for i in range(2):
 printer.completeTask()
printer.scheduleTask('scan 4 pages', 'Scanning')
printer.report()
```
